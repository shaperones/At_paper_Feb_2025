# Binned methylation in original full-length ONSEN copies
## TE body

```python
# Define the number of bins
NUM_BINS = 100

# Initialize a dictionary to store results for each sample and context
results_dict = {}

# Iterate through each sample and context
for sample_context, (df1, onsen1, df2, onsen2) in data_dict.items():
    # Initialize a list to store combined binned methylation for all ONSENs
    combined_binned_methylation = []
    
    # Get unique ONSENs for each barcode
    onsens_barcode1 = onsen1.drop_duplicates().reset_index(drop=True)
    onsens_barcode2 = onsen2.drop_duplicates().reset_index(drop=True)
    
    # Iterate through ONSENs in barcode 1
    for _, onsen_row in onsens_barcode1.iterrows():
        # Extract ONSEN coordinates
        chr_name = onsen_row[0]  # Chromosome
        start = onsen_row[3]     # Start position
        end = onsen_row[4]       # End position
        onsen_name = onsen_row[8].split(';')[0].split('=')[1]  # Extract ONSEN name from column 8
        
        # Initialize lists to store combined methylation values and counts for each bin
        combined_bins_meth = [[] for _ in range(NUM_BINS)]
        combined_bins_count = [0] * NUM_BINS
        
        # Calculate bin size for the current ONSEN
        bin_size = (end - start) / NUM_BINS
        
        # Process methylation data for barcode 1
        onsen_meth_barcode1 = df1[
            (df1[0] == chr_name) &  # Match chromosome
            (df1[1] >= start) &     # Start position >= ONSEN start
            (df1[2] <= end)         # End position <= ONSEN end
        ]
        for _, row in onsen_meth_barcode1.iterrows():
            pos = row[1]  # Methylation position
            bin_idx = min(int((pos - start) // bin_size), NUM_BINS - 1)  # Ensure bin index is within range
            combined_bins_meth[bin_idx].append(row[10])  # Assuming column 10 contains methylation values
            combined_bins_count[bin_idx] += 1  # Count cytosines
        
        # Process methylation data for barcode 2
        onsen_meth_barcode2 = df2[
            (df2[0] == chr_name) &  # Match chromosome
            (df2[1] >= start) &     # Start position >= ONSEN start
            (df2[2] <= end)         # End position <= ONSEN end
        ]
        for _, row in onsen_meth_barcode2.iterrows():
            pos = row[1]  # Methylation position
            bin_idx = min(int((pos - start) // bin_size), NUM_BINS - 1)  # Ensure bin index is within range
            combined_bins_meth[bin_idx].append(row[10])  # Assuming column 10 contains methylation values
            combined_bins_count[bin_idx] += 1  # Count cytosines
        
        # Calculate mean methylation for each bin
        mean_binned_meth = []
        for meth_values, count in zip(combined_bins_meth, combined_bins_count):
            if count > 0:
                mean_meth = np.mean(meth_values)  # Calculate mean methylation
            else:
                mean_meth = 0  # If no cytosines, set methylation to 0
            mean_binned_meth.append(mean_meth)
        
        # Add ONSEN name and mean binned methylation to the results
        combined_binned_methylation.append([onsen_name] + mean_binned_meth)
    
    # Convert the results to a DataFrame
    columns = ['ONSEN'] + [f'bin_{i+1}' for i in range(NUM_BINS)]
    results_df = pd.DataFrame(combined_binned_methylation, columns=columns)
    
    # Store the results in the dictionary
    results_dict[sample_context] = results_df

    # Save the results to CSV files
    results_df.to_csv(f'ONSENs_meth_bins/ONSEN_binned_methylation_{sample_context}.csv', index=False)
    print(sample_context)
```

```python
# Define the bin size and flanking size
BIN_SIZE = 50  # Fixed bin size
FLANKING_SIZE = 2000  # 2000 bp flanking regions

# Initialize a dictionary to store results for each sample and context
results_flank_dict = {}

# Iterate through each sample and context
for sample_context, (df1, onsen1, df2, onsen2) in data_dict.items():
    # Initialize a list to store combined binned methylation for all ONSENs
    combined_binned_methylation = []
    
    # Get unique ONSENs for each barcode
    onsens_barcode1 = onsen1.drop_duplicates().reset_index(drop=True)
    onsens_barcode2 = onsen2.drop_duplicates().reset_index(drop=True)
    
    # Iterate through ONSENs in barcode 1
    for _, onsen_row in onsens_barcode1.iterrows():
        # Extract ONSEN coordinates
        chr_name = onsen_row[0]  # Chromosome
        start = onsen_row[3]     # Start position
        end = onsen_row[4]       # End position
        onsen_name = onsen_row[8].split(';')[0].split('=')[1]  # Extract ONSEN name from column 8
        
        # Define flanking regions
        left_flank_start = start - FLANKING_SIZE
        left_flank_end = start
        right_flank_start = end
        right_flank_end = end + FLANKING_SIZE
        
        # Initialize a list to store methylation values for each bin (for barcode 1)
        onsen_bins_barcode1 = [[] for _ in range((FLANKING_SIZE * 2) // BIN_SIZE)]  # Total bins = 2000 * 2 / 50 = 80
        
        # Process methylation data for barcode 1
        for bin_start in range(left_flank_start, left_flank_end, BIN_SIZE):
            bin_end = bin_start + BIN_SIZE
            # Filter methylation data for the current bin
            bin_data = df1[
                (df1[0] == chr_name) &  # Match chromosome
                (df1[1] >= bin_start) &  # Start position >= bin start
                (df1[2] <= bin_end)      # End position <= bin end
            ]
            # Append methylation values to the corresponding bin
            bin_idx = (bin_start - left_flank_start) // BIN_SIZE
            onsen_bins_barcode1[bin_idx].extend(bin_data[10].tolist())  # Assuming column 10 contains methylation values
        
        for bin_start in range(right_flank_start, right_flank_end, BIN_SIZE):
            bin_end = bin_start + BIN_SIZE
            # Filter methylation data for the current bin
            bin_data = df1[
                (df1[0] == chr_name) &  # Match chromosome
                (df1[1] >= bin_start) &  # Start position >= bin start
                (df1[2] <= bin_end)      # End position <= bin end
            ]
            # Append methylation values to the corresponding bin
            bin_idx = (FLANKING_SIZE // BIN_SIZE) + (bin_start - right_flank_start) // BIN_SIZE
            onsen_bins_barcode1[bin_idx].extend(bin_data[10].tolist())  # Assuming column 10 contains methylation values
        
        # Add ONSEN name and binned methylation (barcode 1) to the results
        combined_binned_methylation.append([onsen_name, onsen_bins_barcode1])
    
    # Iterate through ONSENs in barcode 2
    for _, onsen_row in onsens_barcode2.iterrows():
        # Extract ONSEN coordinates
        chr_name = onsen_row[0]  # Chromosome
        start = onsen_row[3]     # Start position
        end = onsen_row[4]       # End position
        onsen_name = onsen_row[8].split(';')[0].split('=')[1]  # Extract ONSEN name from column 8
        
        # Define flanking regions
        left_flank_start = start - FLANKING_SIZE
        left_flank_end = start
        right_flank_start = end
        right_flank_end = end + FLANKING_SIZE
        
        # Initialize a list to store methylation values for each bin (for barcode 2)
        onsen_bins_barcode2 = [[] for _ in range((FLANKING_SIZE * 2) // BIN_SIZE)]  # Total bins = 2000 * 2 / 50 = 80
        
        # Process methylation data for barcode 2
        for bin_start in range(left_flank_start, left_flank_end, BIN_SIZE):
            bin_end = bin_start + BIN_SIZE
            # Filter methylation data for the current bin
            bin_data = df2[
                (df2[0] == chr_name) &  # Match chromosome
                (df2[1] >= bin_start) &  # Start position >= bin start
                (df2[2] <= bin_end)      # End position <= bin end
            ]
            # Append methylation values to the corresponding bin
            bin_idx = (bin_start - left_flank_start) // BIN_SIZE
            onsen_bins_barcode2[bin_idx].extend(bin_data[10].tolist())  # Assuming column 10 contains methylation values
        
        for bin_start in range(right_flank_start, right_flank_end, BIN_SIZE):
            bin_end = bin_start + BIN_SIZE
            # Filter methylation data for the current bin
            bin_data = df2[
                (df2[0] == chr_name) &  # Match chromosome
                (df2[1] >= bin_start) &  # Start position >= bin start
                (df2[2] <= bin_end)      # End position <= bin end
            ]
            # Append methylation values to the corresponding bin
            bin_idx = (FLANKING_SIZE // BIN_SIZE) + (bin_start - right_flank_start) // BIN_SIZE
            onsen_bins_barcode2[bin_idx].extend(bin_data[10].tolist())  # Assuming column 10 contains methylation values
        
        # Add ONSEN name and binned methylation (barcode 2) to the results
        combined_binned_methylation.append([onsen_name, onsen_bins_barcode2])
    
    # Combine bins from df1 and df2 for each ONSEN and calculate mean methylation
    final_binned_methylation = []
    for onsen_name, bins_barcode1 in combined_binned_methylation:
        # Find corresponding bins for the same ONSEN in barcode 2
        bins_barcode2 = next((bins for name, bins in combined_binned_methylation if name == onsen_name), None)
        
        if bins_barcode2:
            # Combine methylation values bin-wise
            combined_bins = []
            for b1_meth, b2_meth in zip(bins_barcode1, bins_barcode2):
                combined_values = b1_meth + b2_meth  # Combine methylation values from both barcodes
                mean_meth = np.mean(combined_values) if combined_values else 0  # Calculate mean methylation
                combined_bins.append(mean_meth)
            
            # Add ONSEN name and combined binned methylation to the results
            final_binned_methylation.append([onsen_name] + combined_bins)
    
    # Convert the results to a DataFrame
    columns = ['ONSEN'] + [f'bin_{i+1}' for i in range((FLANKING_SIZE * 2) // BIN_SIZE)]
    results_df = pd.DataFrame(final_binned_methylation, columns=columns)
    
    # Store the results in the dictionary
    results_flank_dict[sample_context] = results_df

    # Save the results to CSV files
    results_df.to_csv(f'ONSENs_flanks/ONSEN_flanking_methylation_{sample_context}.csv', index=False)
    print(sample_context)
```

```python
evd24 = pd.read_csv('evd_positions/EVD_barcode24.gff', sep='\t', header=None)
evd01 = pd.read_csv('evd_positions/EVD_barcode01.gff', sep='\t', header=None)
evd25 = pd.read_csv('evd_positions/EVD_barcode25.gff', sep='\t', header=None)
evd29 = pd.read_csv('evd_positions/EVD_barcode29.gff', sep='\t', header=None)
evd19 = pd.read_csv('evd_positions/EVD_barcode19.gff', sep='\t', header=None)
evd20 = pd.read_csv('evd_positions/EVD_barcode20.gff', sep='\t', header=None)
evd21 = pd.read_csv('evd_positions/EVD_barcode21.gff', sep='\t', header=None)
evd22 = pd.read_csv('evd_positions/EVD_barcode22.gff', sep='\t', header=None)
evd23 = pd.read_csv('evd_positions/EVD_barcode23.gff', sep='\t', header=None)
evd26 = pd.read_csv('evd_positions/EVD_barcode26.gff', sep='\t', header=None)
evd27 = pd.read_csv('evd_positions/EVD_barcode27.gff', sep='\t', header=None)
evd28 = pd.read_csv('evd_positions/EVD_barcode28.gff', sep='\t', header=None)
evd30 = pd.read_csv('evd_positions/EVD_barcode30.gff', sep='\t', header=None)
evd31 = pd.read_csv('evd_positions/EVD_barcode31.gff', sep='\t', header=None)
```

```python
data_dict = {
    'Col-0_CG': (meth24_cg, evd24, meth25_cg, evd25),  # CG context, barcodes 24-25
    'Col-0_CHG': (meth24_chg, evd24, meth25_chg, evd25),  # CHG context, barcodes 24-25
    'Col-0_CHH': (meth24_chh, evd24, meth25_chh, evd25),   # CHH context, barcodes 24-25

    '12-1_CG': (meth01_cg, evd01, meth29_cg, evd29),  # CG context, barcodes 01-29
    '12-1_CHG': (meth01_chg, evd01, meth29_chg, evd29),  # CHG context, barcodes 01-29
    '12-1_CHH': (meth01_chh, evd01, meth29_chh, evd29),  # CHH context, barcodes 01-29

    '1-1_CG': (meth19_cg, evd19, meth26_cg, evd26), 
    '1-1_CHG': (meth19_chg, evd19, meth26_chg, evd26),  
    '1-1_CHH': (meth19_chh, evd19, meth26_chh, evd26), 

    '3-6_CG': (meth20_cg, evd20, meth27_cg, evd27), 
    '3-6_CHG': (meth20_chg, evd20, meth27_chg, evd27),  
    '3-6_CHH': (meth20_chh, evd20, meth27_chh, evd27), 

    '7-5_CG': (meth21_cg, evd21, meth28_cg, evd28), 
    '7-5_CHG': (meth21_chg, evd21, meth28_chg, evd28),  
    '7-5_CHH': (meth21_chh, evd21, meth28_chh, evd28), 

    '13-1_CG': (meth22_cg, evd22, meth30_cg, evd30), 
    '13-1_CHG': (meth22_chg, evd22, meth30_chg, evd30),  
    '13-1_CHH': (meth22_chh, evd22, meth30_chh, evd30), 

    '14-2_CG': (meth23_cg, evd23, meth31_cg, evd31), 
    '14-2_CHG': (meth23_chg, evd23, meth31_chg, evd31),  
    '14-2_CHH': (meth23_chh, evd23, meth31_chh, evd31), 
    
}
```

```python
# Define the number of bins
NUM_BINS = 100

# Initialize a dictionary to store results for each sample and context
results_dict = {}

# Iterate through each sample and context
for sample_context, (df1, evd1, df2, evd2) in data_dict.items():
    # Initialize a list to store combined binned methylation for all EVDs
    combined_binned_methylation = []
    
    # Get unique EVDs for each barcode
    evds_barcode1 = evd1.drop_duplicates().reset_index(drop=True)
    evds_barcode2 = evd2.drop_duplicates().reset_index(drop=True)
    
    # Iterate through EVDs in barcode 1
    for _, evd_row in evds_barcode1.iterrows():
        # Extract EVD coordinates
        chr_name = evd_row[0]  # Chromosome
        start = evd_row[3]     # Start position
        end = evd_row[4]       # End position
        evd_name = evd_row[8].split(';')[0].split('=')[1]  # Extract EVD name from column 8
        
        # Initialize lists to store combined methylation values and counts for each bin
        combined_bins_meth = [[] for _ in range(NUM_BINS)]
        combined_bins_count = [0] * NUM_BINS
        
        # Calculate bin size for the current EVD
        bin_size = (end - start) / NUM_BINS
        
        # Process methylation data for barcode 1
        evd_meth_barcode1 = df1[
            (df1[0] == chr_name) &  # Match chromosome
            (df1[1] >= start) &     # Start position >= EVD start
            (df1[2] <= end)         # End position <= EVD end
        ]
        for _, row in evd_meth_barcode1.iterrows():
            pos = row[1]  # Methylation position
            bin_idx = min(int((pos - start) // bin_size), NUM_BINS - 1)  # Ensure bin index is within range
            combined_bins_meth[bin_idx].append(row[10])  # Assuming column 10 contains methylation values
            combined_bins_count[bin_idx] += 1  # Count cytosines
        
        # Process methylation data for barcode 2
        evd_meth_barcode2 = df2[
            (df2[0] == chr_name) &  # Match chromosome
            (df2[1] >= start) &     # Start position >= EVD start
            (df2[2] <= end)         # End position <= EVD end
        ]
        for _, row in evd_meth_barcode2.iterrows():
            pos = row[1]  # Methylation position
            bin_idx = min(int((pos - start) // bin_size), NUM_BINS - 1)  # Ensure bin index is within range
            combined_bins_meth[bin_idx].append(row[10])  # Assuming column 10 contains methylation values
            combined_bins_count[bin_idx] += 1  # Count cytosines
        
        # Calculate mean methylation for each bin
        mean_binned_meth = []
        for meth_values, count in zip(combined_bins_meth, combined_bins_count):
            if count > 0:
                mean_meth = np.mean(meth_values)  # Calculate mean methylation
            else:
                mean_meth = 0  # If no cytosines, set methylation to 0
            mean_binned_meth.append(mean_meth)
        
        # Add EVD name and mean binned methylation to the results
        combined_binned_methylation.append([evd_name] + mean_binned_meth)
    
    # Convert the results to a DataFrame
    columns = ['EVD'] + [f'bin_{i+1}' for i in range(NUM_BINS)]
    results_df = pd.DataFrame(combined_binned_methylation, columns=columns)
    
    # Store the results in the dictionary
    results_dict[sample_context] = results_df

    # Save the results to CSV files
    results_df.to_csv(f'EVD_meth_bins/EVD_binned_methylation_{sample_context}.csv', index=False)
    print(sample_context)
```

## TE flanks

```python
# Define the bin size and flanking size
BIN_SIZE = 50  # Fixed bin size
FLANKING_SIZE = 2000  # 2000 bp flanking regions

# Initialize a dictionary to store results for each sample and context
results_flank_dict = {}

# Iterate through each sample and context
for sample_context, (df1, evd1, df2, evd2) in data_dict.items():
    # Initialize a list to store combined binned methylation for all EVDs
    combined_binned_methylation = []
    
    # Get unique EVDs for each barcode
    evds_barcode1 = evd1.drop_duplicates().reset_index(drop=True)
    evds_barcode2 = evd2.drop_duplicates().reset_index(drop=True)
    
    # Iterate through EVDs in barcode 1
    for _, evd_row in evds_barcode1.iterrows():
        # Extract EVD coordinates
        chr_name = evd_row[0]  # Chromosome
        start = evd_row[3]     # Start position
        end = evd_row[4]       # End position
        evd_name = evd_row[8].split(';')[0].split('=')[1]  # Extract EVD name from column 8
        
        # Define flanking regions
        left_flank_start = start - FLANKING_SIZE
        left_flank_end = start
        right_flank_start = end
        right_flank_end = end + FLANKING_SIZE
        
        # Initialize a list to store methylation values for each bin (for barcode 1)
        evd_bins_barcode1 = [[] for _ in range((FLANKING_SIZE * 2) // BIN_SIZE)]  # Total bins = 2000 * 2 / 50 = 80
        
        # Process methylation data for barcode 1
        for bin_start in range(left_flank_start, left_flank_end, BIN_SIZE):
            bin_end = bin_start + BIN_SIZE
            # Filter methylation data for the current bin
            bin_data = df1[
                (df1[0] == chr_name) &  # Match chromosome
                (df1[1] >= bin_start) &  # Start position >= bin start
                (df1[2] <= bin_end)      # End position <= bin end
            ]
            # Append methylation values to the corresponding bin
            bin_idx = (bin_start - left_flank_start) // BIN_SIZE
            evd_bins_barcode1[bin_idx].extend(bin_data[10].tolist())  # Assuming column 10 contains methylation values
        
        for bin_start in range(right_flank_start, right_flank_end, BIN_SIZE):
            bin_end = bin_start + BIN_SIZE
            # Filter methylation data for the current bin
            bin_data = df1[
                (df1[0] == chr_name) &  # Match chromosome
                (df1[1] >= bin_start) &  # Start position >= bin start
                (df1[2] <= bin_end)      # End position <= bin end
            ]
            # Append methylation values to the corresponding bin
            bin_idx = (FLANKING_SIZE // BIN_SIZE) + (bin_start - right_flank_start) // BIN_SIZE
            evd_bins_barcode1[bin_idx].extend(bin_data[10].tolist())  # Assuming column 10 contains methylation values
        
        # Add EVD name and binned methylation (barcode 1) to the results
        combined_binned_methylation.append([evd_name, evd_bins_barcode1])
    
    # Iterate through EVDs in barcode 2
    for _, evd_row in evds_barcode2.iterrows():
        # Extract EVD coordinates
        chr_name = evd_row[0]  # Chromosome
        start = evd_row[3]     # Start position
        end = evd_row[4]       # End position
        evd_name = evd_row[8].split(';')[0].split('=')[1]  # Extract EVD name from column 8
        
        # Define flanking regions
        left_flank_start = start - FLANKING_SIZE
        left_flank_end = start
        right_flank_start = end
        right_flank_end = end + FLANKING_SIZE
        
        # Initialize a list to store methylation values for each bin (for barcode 2)
        evd_bins_barcode2 = [[] for _ in range((FLANKING_SIZE * 2) // BIN_SIZE)]  # Total bins = 2000 * 2 / 50 = 80
        
        # Process methylation data for barcode 2
        for bin_start in range(left_flank_start, left_flank_end, BIN_SIZE):
            bin_end = bin_start + BIN_SIZE
            # Filter methylation data for the current bin
            bin_data = df2[
                (df2[0] == chr_name) &  # Match chromosome
                (df2[1] >= bin_start) &  # Start position >= bin start
                (df2[2] <= bin_end)      # End position <= bin end
            ]
            # Append methylation values to the corresponding bin
            bin_idx = (bin_start - left_flank_start) // BIN_SIZE
            evd_bins_barcode2[bin_idx].extend(bin_data[10].tolist())  # Assuming column 10 contains methylation values
        
        for bin_start in range(right_flank_start, right_flank_end, BIN_SIZE):
            bin_end = bin_start + BIN_SIZE
            # Filter methylation data for the current bin
            bin_data = df2[
                (df2[0] == chr_name) &  # Match chromosome
                (df2[1] >= bin_start) &  # Start position >= bin start
                (df2[2] <= bin_end)      # End position <= bin end
            ]
            # Append methylation values to the corresponding bin
            bin_idx = (FLANKING_SIZE // BIN_SIZE) + (bin_start - right_flank_start) // BIN_SIZE
            evd_bins_barcode2[bin_idx].extend(bin_data[10].tolist())  # Assuming column 10 contains methylation values
        
        # Add EVD name and binned methylation (barcode 2) to the results
        combined_binned_methylation.append([evd_name, evd_bins_barcode2])
    
    # Combine bins from df1 and df2 for each EVD and calculate mean methylation
    final_binned_methylation = []
    for evd_name, bins_barcode1 in combined_binned_methylation:
        # Find corresponding bins for the same EVD in barcode 2
        bins_barcode2 = next((bins for name, bins in combined_binned_methylation if name == evd_name), None)
        
        if bins_barcode2:
            # Combine methylation values bin-wise
            combined_bins = []
            for b1_meth, b2_meth in zip(bins_barcode1, bins_barcode2):
                combined_values = b1_meth + b2_meth  # Combine methylation values from both barcodes
                mean_meth = np.mean(combined_values) if combined_values else 0  # Calculate mean methylation
                combined_bins.append(mean_meth)
            
            # Add EVD name and combined binned methylation to the results
            final_binned_methylation.append([evd_name] + combined_bins)
    
    # Convert the results to a DataFrame
    columns = ['EVD'] + [f'bin_{i+1}' for i in range((FLANKING_SIZE * 2) // BIN_SIZE)]
    results_df = pd.DataFrame(final_binned_methylation, columns=columns)
    
    # Store the results in the dictionary
    results_flank_dict[sample_context] = results_df

    # Save the results to CSV files
    results_df.to_csv(f'EVDs_flanks/EVD_flanking_methylation_{sample_context}.csv', index=False)
    print(sample_context)
```

```python
# Initialize a list to store all methylation data
all_data = []

# Iterate through each sample and context
for sample_context, (df1_a, df2_a, df1_b, df2_b) in data_dict.items():
    sample, context = sample_context.split('_')  # Split into sample and context
    # Combine data from both barcodes for the same windows
    for df1, df2 in [(df1_a, df2_a), (df1_b, df2_b)]:
        # Iterate through each window in df2
        for index, row in df2.iterrows():
            chr_name = row[0]  # Chromosome name
            left_border = row[3]  # Left border
            right_border = row[4]  # Right border
            attributes = row[8]  # Attributes column
            name = None
            for attr in attributes.split(';'):
                if attr.startswith('Name='):
                    name = attr.split('=')[1]
                    break
            
            # Filter df1 for rows within the current window
            window_data = df1[
                (df1[0] == chr_name) &  # Match chromosome
                (df1[1] >= left_border) &  # Start position >= left border
                (df1[2] <= right_border)  # End position <= right border
            ]
            
            # Extract methylation levels (column 10 in df1)
            methylation_levels = window_data[10]  # Assuming column 10 is index 9 (0-based)
            
            # Append methylation levels, sample, context, and window name to the list
            all_data.extend([{
                'Sample': sample,
                'Context': context,
                'Window': name,
                'Methylation Level': level
            } for level in methylation_levels])

# Create a DataFrame from the combined data
plot_data = pd.DataFrame(all_data)

# Define the order of windows
window_order = onsen01[8].apply(lambda x: x.split('Name=')[1].split(';')[0]).unique()

# Initialize the figure with subplots for each window and context
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 6))
fig.suptitle('Methylation Level Distribution by ONSEN, Context, and Sample', fontsize=20, y=1.02)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Iterate through each window and context
for i, window in enumerate(['AT3TE92525', 'AT1TE12295']):
    for j, context in enumerate(['CG', 'CHG', 'CHH']):
        ax = axes[i*3 + j]  # Calculate the subplot index
        # Filter data for the current window and context
        window_context_data = plot_data[(plot_data['Window'] == window) & (plot_data['Context'] == context)]
        
        # Draw the violin plot
        sns.violinplot(
            x='Sample', y='Methylation Level', data=window_context_data,
            ax=ax, palette=colors, cut=0
        )
        
        # Add labels and title for the subplot
        ax.set_title(f'{window} ({context})', fontsize=16, pad=10)
        ax.set_xlabel('')
        ax.set_ylabel('Methylation Level (%)', fontsize=14, labelpad=8)
        ax.set_ylim(0, 100)  # Set y-axis limits to 0-100
        
        # Customize tick labels
        ax.tick_params(axis='both', which='both', labelsize=13, length=0)  # Hide tick marks
        
        # Rotate x-axis labels for better readability
        ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha='right')
        
        # Remove spines (top and right borders)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)

plt.tight_layout()  # Prevent overlap
plt.savefig('plots/violin_by_onsen_context_sample_AT1TE12295_AT3TE92525.svg', dpi=300)
```

```python
def calculate_stats(methylation_levels):
    return np.mean(methylation_levels), np.median(methylation_levels), np.std(methylation_levels)

def perform_tests(group1, group2):
    t_stat, t_pvalue = ttest_ind(group1, group2)
    ks_stat, ks_pvalue = ks_2samp(group1, group2)
   
    return t_pvalue, ks_pvalue


# Define the data for barcodes and ONSENs for all contexts (CG, CHG, CHH)
barcode_data = [
    # CG context
    (meth01_cg, onsen01),  # Barcode 01
    (meth29_cg, onsen29),  # Barcode 29
    (meth24_cg, onsen24),  # Barcode 24
    (meth25_cg, onsen25),  # Barcode 25
    # CHG context
    (meth01_chg, onsen01),  # Barcode 01
    (meth29_chg, onsen29),  # Barcode 29
    (meth24_chg, onsen24),  # Barcode 24
    (meth25_chg, onsen25),  # Barcode 25
    # CHH context
    (meth01_chh, onsen01),  # Barcode 01
    (meth29_chh, onsen29),  # Barcode 29
    (meth24_chh, onsen24),  # Barcode 24
    (meth25_chh, onsen25)   # Barcode 25
]

# Initialize a list to store results
results = []

# Iterate through each window (assuming window names are the same across ONSENs)
# Use the first ONSEN (onsen01) to get window names
for index, row in onsen01.iterrows():
    # Extract window information
    chr_name = row[0]  # Chromosome name
    attributes = row[8]  # Attributes column
    name = None
    for attr in attributes.split(';'):
        if attr.startswith('Name='):
            name = attr.split('=')[1]
            break
    
    # Initialize lists to store methylation levels for 01-29 and 24-25 for each context
    methylation_01_29_cg = []
    methylation_24_25_cg = []
    methylation_01_29_chg = []
    methylation_24_25_chg = []
    methylation_01_29_chh = []
    methylation_24_25_chh = []
    
    # Extract methylation levels for the current window from each barcode and context
    for df1, df2 in barcode_data:
        # Find the corresponding row in df2 (ONSEN data) for the current window
        window_row = df2[df2[8].str.contains(f'Name={name}')].iloc[0]  # Match by Name
        left_border = window_row[3]  # Left border from df2
        right_border = window_row[4]  # Right border from df2
        
        # Filter df1 for rows within the current window
        window_data = df1[
            (df1[0] == chr_name) &  # Match chromosome
            (df1[1] >= left_border) &  # Start position >= left border
            (df1[2] <= right_border)  # End position <= right border
        ]
        
        # Extract methylation levels (column 10 in df1)
        methylation_levels = window_data[10]  # Assuming column 10 is index 9 (0-based)
        
        # Append methylation levels to the appropriate list based on context and barcode
        if df2.equals(onsen01) or df2.equals(onsen29):
            if df1.equals(meth01_cg) or df1.equals(meth29_cg):
                methylation_01_29_cg.extend(methylation_levels)
            elif df1.equals(meth01_chg) or df1.equals(meth29_chg):
                methylation_01_29_chg.extend(methylation_levels)
            elif df1.equals(meth01_chh) or df1.equals(meth29_chh):
                methylation_01_29_chh.extend(methylation_levels)
        elif df2.equals(onsen24) or df2.equals(onsen25):
            if df1.equals(meth24_cg) or df1.equals(meth25_cg):
                methylation_24_25_cg.extend(methylation_levels)
            elif df1.equals(meth24_chg) or df1.equals(meth25_chg):
                methylation_24_25_chg.extend(methylation_levels)
            elif df1.equals(meth24_chh) or df1.equals(meth25_chh):
                methylation_24_25_chh.extend(methylation_levels)
    
    # Calculate summary statistics for each context
    
    
    mean_01_29_cg, median_01_29_cg, std_01_29_cg = calculate_stats(methylation_01_29_cg)
    mean_24_25_cg, median_24_25_cg, std_24_25_cg = calculate_stats(methylation_24_25_cg)
    mean_01_29_chg, median_01_29_chg, std_01_29_chg = calculate_stats(methylation_01_29_chg)
    mean_24_25_chg, median_24_25_chg, std_24_25_chg = calculate_stats(methylation_24_25_chg)
    mean_01_29_chh, median_01_29_chh, std_01_29_chh = calculate_stats(methylation_01_29_chh)
    mean_24_25_chh, median_24_25_chh, std_24_25_chh = calculate_stats(methylation_24_25_chh)
    
    # Perform statistical tests for each context
    
    t_pvalue_cg, ks_pvalue_cg = perform_tests(methylation_01_29_cg, methylation_24_25_cg)
    t_pvalue_chg, ks_pvalue_chg = perform_tests(methylation_01_29_chg, methylation_24_25_chg)
    t_pvalue_chh, ks_pvalue_chh = perform_tests(methylation_01_29_chh, methylation_24_25_chh)
    
    results.append({
        'Window': name,
        'Context': 'CG',
        'Mean (01-29)': mean_01_29_cg,
        'Median (01-29)': median_01_29_cg,
        'Std (01-29)': std_01_29_cg,
        'Mean (24-25)': mean_24_25_cg,
        'Median (24-25)': median_24_25_cg,
        'Std (24-25)': std_24_25_cg,
        'T-test p-value': t_pvalue_cg,
        'KS-test p-value': ks_pvalue_cg,
    })
    results.append({
        'Window': name,
        'Context': 'CHG',
        'Mean (01-29)': mean_01_29_chg,
        'Median (01-29)': median_01_29_chg,
        'Std (01-29)': std_01_29_chg,
        'Mean (24-25)': mean_24_25_chg,
        'Median (24-25)': median_24_25_chg,
        'Std (24-25)': std_24_25_chg,
        'T-test p-value': t_pvalue_chg,
        'KS-test p-value': ks_pvalue_chg,
    })
    results.append({
        'Window': name,
        'Context': 'CHH',
        'Mean (01-29)': mean_01_29_chh,
        'Median (01-29)': median_01_29_chh,
        'Std (01-29)': std_01_29_chh,
        'Mean (24-25)': mean_24_25_chh,
        'Median (24-25)': median_24_25_chh,
        'Std (24-25)': std_24_25_chh,
        'T-test p-value': t_pvalue_chh,
        'KS-test p-value': ks_pvalue_chh,
    })

results_df = pd.DataFrame(results)

#results_df.to_csv('methylation_comparison_results_all_contexts.csv', index=False)

results_df
```
